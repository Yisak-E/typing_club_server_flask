# copilot generated animations

typing animation

```python
import tkinter as tk

class TypingAnimation:
    def __init__(self, root):
        self.root = root
        self.text = "Typing practice"
        self.index = 0

        # Label to show typing text
        self.label = tk.Label(root, text="", font=("Arial", 24), fg="blue")
        self.label.pack(pady=50)

        # Start animation
        self.type_next_char()

    def type_next_char(self):
        if self.index < len(self.text):
            # Add next character
            current_text = self.label.cget("text")
            self.label.config(text=current_text + self.text[self.index])
            self.index += 1

            # Schedule next character after 100ms (0.1 second)
            self.root.after(100, self.type_next_char)


# Main window
root = tk.Tk()
root.title("Typing Animation")
root.geometry("600x200")

typing = TypingAnimation(root)

root.mainloop()
```

## square rotating anime

```python
import tkinter as tk
import math

class RotatingSquare:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=400, height=400, bg="white")
        self.canvas.pack()

        # Square parameters
        self.center_x = 200
        self.center_y = 200
        self.size = 100
        self.angle = 0

        # Initial square
        self.square = self.create_square(self.angle)

        # Start animation
        self.rotate()

    def create_square(self, angle):
        """Return coordinates of a rotated square centered at (center_x, center_y)."""
        half = self.size / 2
        # Original corners relative to center
        corners = [
            (-half, -half),
            (half, -half),
            (half, half),
            (-half, half)
        ]

        rotated = []
        for x, y in corners:
            # Apply rotation
            xr = x * math.cos(angle) - y * math.sin(angle)
            yr = x * math.sin(angle) + y * math.cos(angle)
            rotated.append((self.center_x + xr, self.center_y + yr))

        # Flatten list of tuples
        coords = [coord for point in rotated for coord in point]
        return coords

    def rotate(self):
        # Increase angle
        self.angle += math.radians(5)  # rotate 5 degrees per frame

        # Update square coordinates
        coords = self.create_square(self.angle)
        self.canvas.delete("square")
        self.canvas.create_polygon(coords, fill="black", tags="square")

        # Schedule next frame
        self.root.after(50, self.rotate)  # 20 frames per second


# Main window
root = tk.Tk()
root.title("Rotating Square")
root.geometry("400x400")

RotatingSquare(root)

root.mainloop()
```

## circle growing anim


```python
import tkinter as tk

class GrowingCircle:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=400, height=400, bg="white")
        self.canvas.pack()

        # Circle parameters
        self.center_x = 200
        self.center_y = 200
        self.radius = 20
        self.target_radius = 150
        self.steps = 50
        self.step_time = 50  # ms per frame
        self.dr = (self.target_radius - self.radius) / self.steps

        # Draw initial circle
        self.circle = self.canvas.create_oval(
            self.center_x - self.radius,
            self.center_y - self.radius,
            self.center_x + self.radius,
            self.center_y + self.radius,
            fill="black"
        )

        # Start animation
        self.animate_step(0)

    def animate_step(self, step):
        if step <= self.steps:
            self.radius += self.dr
            # Update circle coordinates
            self.canvas.coords(
                self.circle,
                self.center_x - self.radius,
                self.center_y - self.radius,
                self.center_x + self.radius,
                self.center_y + self.radius
            )
            # Schedule next frame
            self.root.after(self.step_time, lambda: self.animate_step(step + 1))


# Main window
root = tk.Tk()
root.title("Growing Circle Animation")
root.geometry("400x400")

GrowingCircle(root)

root.mainloop()
```

## shrinking

```python
import tkinter as tk

class ShrinkingCircle:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=400, height=400, bg="white")
        self.canvas.pack()

        # Circle parameters
        self.center_x = 200
        self.center_y = 200
        self.radius = 150          # start large
        self.target_radius = 20    # shrink to small
        self.steps = 50
        self.step_time = 50  # ms per frame
        self.dr = (self.target_radius - self.radius) / self.steps

        # Draw initial circle
        self.circle = self.canvas.create_oval(
            self.center_x - self.radius,
            self.center_y - self.radius,
            self.center_x + self.radius,
            self.center_y + self.radius,
            fill="black"
        )

        # Start animation
        self.animate_step(0)

    def animate_step(self, step):
        if step <= self.steps:
            self.radius += self.dr
            # Update circle coordinates
            self.canvas.coords(
                self.circle,
                self.center_x - self.radius,
                self.center_y - self.radius,
                self.center_x + self.radius,
                self.center_y + self.radius
            )
            # Schedule next frame
            self.root.after(self.step_time, lambda: self.animate_step(step + 1))


# Main window
root = tk.Tk()
root.title("Shrinking Circle Animation")
root.geometry("400x400")

ShrinkingCircle(root)

root.mainloop()
```

## bouncing balls


```python
import tkinter as tk

class BouncingBall:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=400, height=400, bg="white")
        self.canvas.pack()
        self.ball = self.canvas.create_oval(180, 180, 220, 220, fill="red")
        self.dx, self.dy = 4, 3
        self.animate()

    def animate(self):
        self.canvas.move(self.ball, self.dx, self.dy)
        x1, y1, x2, y2 = self.canvas.coords(self.ball)
        if x1 <= 0 or x2 >= 400: self.dx = -self.dx
        if y1 <= 0 or y2 >= 400: self.dy = -self.dy
        self.canvas.after(30, self.animate)

root = tk.Tk()
BouncingBall(root)
root.mainloop()
```

## color fade

```python
import tkinter as tk

class ColorFade:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=400, height=200, bg="white")
        self.canvas.pack()
        self.rect = self.canvas.create_rectangle(50, 50, 350, 150, fill="black")
        self.step = 0
        self.fade()

    def fade(self):
        if self.step <= 255:
            color = f"#{self.step:02x}{self.step:02x}{255-self.step:02x}"
            self.canvas.itemconfig(self.rect, fill=color)
            self.step += 5
            self.canvas.after(50, self.fade)

root = tk.Tk()
ColorFade(root)
root.mainloop()
```

## Sliding text

```python
import tkinter as tk

class SlidingText:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=400, height=100, bg="white")
        self.canvas.pack()
        self.text = self.canvas.create_text(400, 50, text="Hello Tkinter!", font=("Arial", 20), fill="blue")
        self.animate()

    def animate(self):
        self.canvas.move(self.text, -5, 0)
        x, y = self.canvas.coords(self.text)
        if x < 0:  # reset to right side
            self.canvas.coords(self.text, 400, y)
        self.canvas.after(50, self.animate)

root = tk.Tk()
SlidingText(root)
root.mainloop()
```

## pulse effect

```python
import tkinter as tk

class PulseCircle:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=400, height=400, bg="white")
        self.canvas.pack()
        self.radius = 50
        self.growing = True
        self.circle = self.canvas.create_oval(150, 150, 250, 250, fill="purple")
        self.animate()

    def animate(self):
        if self.growing:
            self.radius += 2
            if self.radius >= 100: self.growing = False
        else:
            self.radius -= 2
            if self.radius <= 50: self.growing = True

        self.canvas.coords(self.circle,
            200 - self.radius, 200 - self.radius,
            200 + self.radius, 200 + self.radius
        )
        self.canvas.after(50, self.animate)

root = tk.Tk()
PulseCircle(root)
root.mainloop()
```
